```python
def rename(file_dir):
    file_list = glob.glob(file_dir+'*')
    file_list.sort()
    for f in file_list:
        new_name = f[:-4]+'.csv'
        os.rename(f, new_name)
```

```python
def read_info(odir):
    attr_list=["Collection_Name", "Site_Name", "Location", "Northernmost_Latitude", "Westernmost_Longitude", "Elevation", "Earliest_Year", "Most_Recent_Year", "Species_Name", "Common_Name"]
    df = pd.DataFrame(columns=attr_list)
    file_list = glob.glob(odir+'*-noaa.csv')
    file_list.sort()
    for f in file_list:
        info = []
        with open(f,'r') as file:
            lines = file.readlines()

            for attr in attr_list:
                for l in lines:
                    attr_values = np.nan
                    words = l.split()
                    #print(words)
                    if len(words) >= 3:
                        #print(words[1][:-1])
                        if words[1][:-1] == attr:
                            attr_values = words[2:]
                            attr_values = ' '.join(attr_values) 
                            break
                info.append(attr_values)
        print(info)
        df.loc[len(df.index)] = info
    return df
```

```python
def read_info(fdir,odir):
    #import the list of files directories
    file_list = glob.glob(fdir+'*-noaa.csv')
    file_list.sort()
    #loop through the file list
    for f in file_list:
        #get the name of the file
        fname = os.path.basename(f)
        print(fname)
        lists = []
        #open the file
        with open(f,'r') as file:
            #read by lines into a list
            lines = file.readlines()
            for l in lines:
                #if the row starts with a #, we consider it as a line in the description
                if l[0] == "#":
                    continue
                else:
                    lists.append(l)
        #create a new pandas dataframe
        df = pd.DataFrame()
        #split the rows into segments by its delimiter       
        for i in range(len(lists)):
            if '\t' in lists[i]:
                df[i] = lists[i].split('\t')
        #append the elements into the created dataframe and output to a csv file
        df.iloc[:-1,:].T.to_csv(odir+fname,index=False, header=False)
        del df, lists
```
